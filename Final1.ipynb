```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("train.csv")
df.head()
sns.countplot(x='Survived', color='red',data=df)
plt.show()
sns.countplot(x='Pclass', hue='Survived',color='yellow',data=df)
plt.title("Passenger Class")
plt.show()
sns.countplot(x='Sex', hue='Survived',color='gold',data=df)
plt.title("Survival Gender")
plt.show()
plt.figure(figsize=(10,5))
sns.histplot(df[df['Survived']==1]['Age'], bins=30, kde=True, color='lime',label='Survived')
sns.histplot(df[df['Survived']==0]['Age'], bins=30, kde=True, color='violet',label='Not Survived')
plt.legend()
plt.title("Survived Distribution by Age")
plt.show()
survived_counts=df['Survived'].value_counts()
plt.figure()
plt.pie(survived_counts,
        labels=['Not Survival', 'Survival'],
        autopct='%1.2f%%',
        startangle=90)
plt.title("Survival Rate")
plt.show()
gender_survived=df.groupby('Sex')['Survived'].mean()
plt.figure()
plt.pie(gender_survived,
        labels=gender_survived.index,
        autopct='%2.2f%%',
        startangle=90)
plt.title("Survival Gender")
plt.show

survival=df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(10,5))
plt.plot(survival.index, survival.values,marker='o',markersize=10,color='black',markerfacecolor='maroon')
plt.xlabel('Passanger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Class Passanger')
plt.xticks([1,2,3])
plt.show()
```
